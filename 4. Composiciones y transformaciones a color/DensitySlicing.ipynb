```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Density slicing</h2>
<p>Se realizará un <i>density slicing</i> sobre una imagen monobanda. Elegimos al menos 4 rangos de ND y aplicamos colores de nuestra elección. Comprobaremos los resultados sobre la imagen NDVI obtenida en el ejercicio anterior.</p>

```python
def densi(img):
    densi_res=np.zeros((img.shape[0], img.shape[1], 3), np.uint8)
    for x in range(densi_res.shape[0]):
        for y in range(densi_res.shape[1]):
            if 50<img[x,y]<=130:
                densi_res[x,y]=[127,127,127] #gris
            elif 130<img[x,y]<=140:
                densi_res[x,y]=[102,204,255]
            elif 140<img[x,y]<=145:
                densi_res[x,y]=[102,204,255]
            elif 145<img[x,y]<=153:
                densi_res[x,y]=[0,153,153]
            elif 153<img[x,y]<=160:
                densi_res[x,y]=[0,153,153]
            elif 160<img[x,y]<=170:
                densi_res[x,y]=[51,51,204]
            elif 170<img[x,y]<=176:
                densi_res[x,y]=[51,51,204]
            elif 176<img[x,y]<=185:
                densi_res[x,y]=[51,51,204]
            elif 185<img[x,y]<=195:
                densi_res[x,y]=[51,51,204]
    
    return densi_res
```

```python
img_ndvi=cv2.imread('modificadas/Ej3_NDVI.png', cv2.IMREAD_GRAYSCALE)
density_slicing=densi(img_ndvi)
img_ds=cv2.cvtColor(density_slicing, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10, 10))
plt.imshow(img_ds)
plt.show()
```

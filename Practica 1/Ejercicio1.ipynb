```python
"""Instrucciones básicas del lenguaje.
Creación de arrays (listas) de valores en python."""

import numpy

x=numpy.arange(0,10,1)  #Generamos un array de numpy de 0 a 10 con paso de 1. Nota: el último valor del rango no se añade por defecto.
print(x)
y=x**2 #Generamos el array y cuyo valor será el cuadrado de los valores de x.
print(y)
z=x*2  #Generamos el array z cuyo valor será el doble de los valores de x.
print(z)

c=numpy.arange(10,21,2)  #Generamos un array de numpy de 10 a 21 con paso de 2. Nota: El último valor del rango no se añade por defecto.
print(c)
d=c-1  #Calculamos el array d con valores iguales a los del array c-1.
print(d)
```

```python
print(numpy.sin(numpy.pi/4)) #Podemos calcular los valores del seno de pi/4 a partir de la librería de numpy.

x=numpy.arange(0,numpy.pi,0.01)  #Generamos un array de 0 a pi con un paso de 0.01
print(x)

y=numpy.sin(x)  #Calculamos todos los valores del seno del array x.
print(y[0],"...",y[314])

z=numpy.cos(x)  #Calculamos todos los valores del coseno del array x
print(z[0],"...",z[314])
```

```python
"""Funciones para dibujar gráficas."""
#Funciones extra para las gráficas (zoom,pan,etc...)
%matplotlib notebook
import matplotlib.pyplot as plt

plt.figure(1)
plt.plot(x)
plt.ylabel('Lista de valores de x')
plt.show()
```

```python
plt.figure(2)
plt.plot(y)
plt.ylabel('Lista de valores de y')
```

```python
plt.figure(3)
plt.plot(x,y)
plt.ylabel('Sin(x)')
plt.ylabel('x')
```

```python
plt.figure(4)
plt.plot(x,y,'r')
plt.plot(x,z,'g')
plt.legend(['Sin(x)','Cos(x)'])
plt.xlabel('x')
plt.ylabel('Values')
```

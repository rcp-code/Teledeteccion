```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

<h2>Ejercicio 5:</h2>
<p>Escriba una función que realice un corte de colas en HSV (cortehsv) a una determinada composición en color. Para ello, elija una combinación en color verdadero a partir de las imágenes disponibles y siga los pasos descritos a continuación:</p>
<p>• Transforme la combinación (RGB) al dominio HSV.</p>
<p>• Realice un corte de colas de la componente S o de la componente V.</p>
<p>• Transforme la imagen resultante (HSV) de nuevo al dominio RGB (ejemplo en la plantilla).</p>

```python
#Recupero la función corte del ejercicio 1:
def corte(img, porc):
    resimg=img.copy()
    #Histograma de la imagen:
    h={x:0 for x in range(0,256)}
    for x in img:
        for y in x:
            h[y]=h[y]+1
    l_valores=list(h.values())  
    num_pixeles=img.size
    porct=(num_pixeles*porc)
    
    #Obtención de los cortes mínimo y máximo
    cmin=0
    cmax=255
    v=0
    for x in range(1,256):
        v+=l_valores[x]
        if v>porct:
            cmin=x
            break
    v=0
    for x in range(255,0,-1):  #recorrido hacia atrás para encontrar el máximo
        v+=l_valores[x]
        if v>porct:
            cmax=x
            break

    #Se aplica el corte:
    for x in range(resimg.shape[0]):         
        for y in range(resimg.shape[1]):     
            if resimg[x][y]<=cmin:    #todos los valores menores que cmin se ponen a 0 
                resimg[x][y]=0               
            elif resimg[x][y]>=cmax:  #todos los valores mayores o iguales que cmax se ponen a 255       
                resimg[x][y]=255             
            else:                            
                resimg[x][y]=0+((resimg[x][y]-cmin)/(cmax-cmin))*(255-0)
    
    return resimg
```

```python
def cortehsv(img):
    resimg=img.copy()
    hsv=cv2.cvtColor(resimg,cv2.COLOR_BGR2HSV) #transformación al dominio HSV
    h,s,v=cv2.split(hsv)
    corteS=corte(s, 0.01)  #Corte de colas de la componente S
    corteV=corte(v, 0.01)  #Corte de colas de la componente V
    resimgS=cv2.merge((h,corteS,v))
    resimgV=cv2.merge((h,s,corteV))
    resimgSV=cv2.merge((h,corteS,corteV))
    #Transformación de HSV al dominio RGB
    resimgS=cv2.cvtColor(resimgS,cv2.COLOR_HSV2BGR)
    resimgV=cv2.cvtColor(resimgV,cv2.COLOR_HSV2BGR)
    resimgSV=cv2.cvtColor(resimgSV,cv2.COLOR_HSV2BGR)
    
    return resimgS,resimgV,resimgSV
```

```python
img_CV=cv2.imread('modificadas/Ej1_Color_verdadero.png', cv2.IMREAD_COLOR)
cs,cv,csv=cortehsv(img_CV)
original=cv2.cvtColor(img_CV, cv2.COLOR_BGR2RGB)
img_cs=cv2.cvtColor(cs, cv2.COLOR_BGR2RGB)
img_cv=cv2.cvtColor(cv, cv2.COLOR_BGR2RGB)
img_csv=cv2.cvtColor(csv, cv2.COLOR_BGR2RGB)

plt.figure(1, figsize=(7, 7))
plt.imshow(original)
plt.title("Color Verdadero", fontsize=20)

plt.figure(2, figsize=(7, 7))
plt.imshow(img_cs)
plt.title("Corte de colas de la componente S (0.01) en dominio HSV de la imagen en color verdadero", fontsize=10)

plt.figure(3, figsize=(7, 7))
plt.imshow(img_cv)
plt.title("Corte de colas de la componente V (0.01) en dominio HSV de la imagen en color verdadero", fontsize=10)

plt.figure(4, figsize=(7, 7))
plt.imshow(img_csv)
plt.title("Corte de colas de la componente S (0.01) y V (0.01) en dominio HSV de la imagen en color verdadero", fontsize=10)

plt.show()
```

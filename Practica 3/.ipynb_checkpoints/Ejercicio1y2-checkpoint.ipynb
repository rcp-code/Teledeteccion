Escriba una función llamada filtro que realice el filtrado lineal de una imagen monobanda a partir de una matriz 3x3 de coeficientes de filtrado indicada como parámetro de entrada. La función debe devolver la imagen resultante así como visualizarla en pantalla.

```python
def filtro(img, cf):
    resimg=img.copy()
    
    #Como la matriz de coeficientes de filtrado se especifica que será de 3x3 podemos
    #suponer que fc=cc=1. Lo mismo para offsetF y offsetC:
    fc=1 #fila central de la matriz de coeficientes cf
    cc=1 #columna central de la matriz de coeficientes cf
    offsetF=1
    offsetC=1
    
    for f in range(resimg.shape[0]):
        for c in range(resimg.shape[0]):
            if f==0 or f==resimg.shape[0]-1:
                resimg[f][c]=0   #Se rellena con ceros los bordes de la matriz de la imagen
            elif c==0 or c==resimg.shape[0]-1:
                resimg[f][c]=0  #Se rellena con ceros los bordes de la matriz de la imagen
            else:
                resimg[f][c]=np.rint(suma_ND(img, f, c, cf, fc, cc, offsetF, offsetC)/suma_CF(cf, fc, cc, offsetF, offsetC))
    return resimg

#Función auxiliar para calcular el sumatorio de los niveles digitales con la matriz de coeficientes de filtrado:
def suma_ND(img, fila, col, cf, fc, cc, offsetF, offsetC):
    sumatorioND=0
    for i in range(-offsetF,offsetF+1):
        for j in range(-offsetC,offsetC+1):
            sumatorioND+=img[fila+i][col+j]*cf[fc+i][cc+j]
    return sumatorioND

#Función auxiliar para calcular el denominador con los coeficientes de la matriz de filtrado
def suma_CF(cf, fc, cc, offsetF, offsetC):
    sumatorioCF=0
    for i in range(-offsetF,offsetF+1):
        for j in range(-offsetC,offsetC+1):
            sumatorioCF+=cf[fc+i][cc+j]
    return sumatorioCF
```

```python
print(filtro(img,cf))
```

Ejercicio 2: A continuación se prueba la función anterior con tres filtros diferentes: media, media+imagen y gaussiano.

```python
#filtros:
filtro_med=np.array([[1,1,1],[1,1,1],[1,1,1]])
filtro_med_img=np.array([[1,1,1],[1,2,1],[1,1,1]])
filtro_gauss=np.array([[1,2,1],[2,4,2],[1,2,1]])
```

```python
img=cv2.imread('../imgpack/Monobanda/7_Lake Erie.jpg',cv2.IMREAD_GRAYSCALE)
img_exp1=filtro(img,filtro_med)
img_exp2=filtro(img,filtro_med_img)
img_exp3=filtro(img,filtro_gauss)
```

```python
fig,axs=plt.subplots(2,2)
axs[0,0].imshow(img, cmap='gray')
axs[0,1].imshow(img_exp1, cmap='gray')
axs[1,0].imshow(img_exp2, cmap='gray')
axs[1,1].imshow(img_exp3, cmap='gray')
```

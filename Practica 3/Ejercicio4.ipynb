```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from math import sqrt
```

```python
def sobel(img):
    resimg=img.copy()
    for fila in range(resimg.shape[0]):
        for col in range(resimg.shape[0]):
            if fila==0 or fila==resimg.shape[0]-1:
                resimg[fila][col]=0   #Se rellena con ceros los bordes de la matriz de la imagen
            elif col==0 or col==resimg.shape[0]-1:
                resimg[fila][col]=0  #Se rellena con ceros los bordes de la matriz de la imagen
            else:
                c=calculo_C(img,fila,col)
                f=calculo_F(img,fila,col)
                resimg[fila][col]=sqrt((c**2)+(f**2))
    return resimg

#Funciones auxiliares que calculan C y F para hallar el pixel central de la ventana 3x3
def calculo_C(img,f,c):
    return (img[f-1][c+1]+2*img[f][c+1]+img[f+1][c+1])-(img[f-1][c-1]+2*img[f][c-1]+img[f+1][c-1])

def calculo_F(img,f,c):
    return (img[f-1][c-1]+2*img[f-1][c]+img[f-1][c+1])-(img[f+1][c-1]+2*img[f+1][c]+img[f+1][c+1])
```

```python
#Test sobre la matriz de teor√≠a
img=np.array([[12,14,17,24,32,30],
            [19,18,17,26,36,40],
            [26,21,17,28,40,42],
            [20,19,18,26,33,35],
            [14,17,20,23,27,20],
            [35,14,22,13,40,21]])
print(sobel(img))
```

```python
img=cv2.imread('../imgpack/Monobanda/7_Lake Erie.jpg',cv2.IMREAD_GRAYSCALE)
img_sobel=sobel(img)
plt.figure(1)
plt.imshow(img,cmap='gray')
plt.figure(2)
plt.imshow(img_sobel,cmap='gray')
```

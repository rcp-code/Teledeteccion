```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def Histograma(imagen):
    h={x:0 for x in range(256)}  #Diccionario generado con valores de x dentro del rango [0, 255] y con valores 0
    for i in imagen:
        for v in i:
            h[v]=h[v]+1   #se introducen los nuevos valores en el histograma
    return list(h.values())       #devuelve el histograma
```

```python
def ecual(img):
    resimg=img.copy()  #copiar imagen
    histograma=Histograma(img)              #calcula el histograma de la imagen original
    fac=np.cumsum(histograma)               #frecuencia acumulada de cada ND en la imagen original (usando cumsum: suma acumulada)
    fe=255/img.size                         #factor de escala
    for x in range(len(fac)):
        fac[x]=np.rint(fac[x]*fe)         #la nueva frecuencia redondeada al entero m√°s cercano
    for x in range(resimg.shape[0]):        
        for y in range(resimg.shape[0]):
            resimg[x,y]=fac[resimg[x,y]]  #
    return resimg
```

```python
img=cv2.imread('../imgpack/Monobanda/7_Lake Erie.jpg',cv2.IMREAD_GRAYSCALE)
#Aplica el corte de colas:
img_exp=ecual(img)
#Dibuja el histograma entre 0 y 254:
plt.hist(img_exp.ravel(),255,[0,254],width=1)
plt.show()
plt.figure(1)
plt.imshow(img,cmap='gray')
plt.figure(2)
plt.imshow(img_exp,cmap='gray')
```
